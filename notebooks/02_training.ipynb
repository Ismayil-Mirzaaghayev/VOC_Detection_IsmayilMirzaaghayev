This notebook trains a YOLOv8n object detection model using the prepared dataset. The model is initialized with COCO pretrained weights and fine-tuned on the annotated images. Training is performed with a resolution of 640×640, batch size of 16, and a fixed random seed of 42 for reproducibility. Hyperparameters such as learning rate, optimizer, early stopping, and data augmentation are specified and justified as part of the training configuration.

```python
from ultralytics import YOLO
import random
import numpy as np
import torch
from pathlib import Path

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

print("Seed set to", SEED)
```

```python
data_yaml = Path("../data/data.yaml").resolve()
output_project = Path("../runs").resolve()

print("data.yaml path:", data_yaml)
print("data.yaml exists:", data_yaml.exists())
print("output project path:", output_project)
```

```python
model = YOLO("yolov8n.pt")

results = model.train(
    data=str(data_yaml),
    epochs=50,
    imgsz=640,
    batch=16,
    seed=SEED,
    lr0=0.01,
    optimizer="SGD",
    patience=15,
    mosaic=1.0,
    project=str(output_project),
    name="yolov8n_exp1"
)
```
