This notebook evaluates the trained YOLOv8n model on the held-out test dataset. It reports detection performance using metrics such as mAP@0.5, mAP@0.5:0.95, precision, recall, and F1 score for each class. The notebook also measures inference speed and visualizes model predictions by displaying bounding boxes on test images, providing insights into the strengths and limitations of the trained detector.

```python
from ultralytics import YOLO
from pathlib import Path
```

```python
weights = Path("../runs/yolov8n_exp1/weights/best.pt").resolve()
data_yaml = Path("../data/data.yaml").resolve()

print("Weights path:", weights)
print("Weights exist:", weights.exists())

print("Dataset yaml:", data_yaml)
print("Dataset yaml exists:", data_yaml.exists())
```

```python
model = YOLO(str(weights))
```

```python
metrics = model.val(
    data=str(data_yaml),
    split="test",
    imgsz=640,
    device="cpu"
)

print(metrics)
```

```python
print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```

```python
test_images = Path("../data/images/test")

model.predict(
    source=str(test_images),
    save=True,
    imgsz=640,
    conf=0.25,
    device="cpu"
)
```
